Keywords rules have the same mapping used for feature engineering. Save to JSON for cleaner Python scripts later.

```python
with open('../app/keywords.json') as f:
    keyword_rules = json.load(f)
    
print(json.dumps(keyword_rules, indent=4))
```

```python
keyword_rules['software_engineer']
```

```python
def parse_job_description(description: str) -> dict:
    description = description.lower()

    extracted_features = {
        'location': 'Central',
        'classifications/0/sub': 'Help Desk & IT Support',
        'workTypes/0': 'Full-time'
    }

    for feature, keywords in keyword_rules.items():
        extracted_features[feature] = int(bool(re.search(keywords, description)))

    return extracted_features

features = parse_job_description(sample_description)
features
```
