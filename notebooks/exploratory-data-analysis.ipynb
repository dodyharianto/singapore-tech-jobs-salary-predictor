```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/sg-tech-jobs-with-salary.csv')
df
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(x='salary_lower', data=df, kde=True)
plt.axvline(df['salary_lower'].median(), color='red', linestyle='--', label='Median')
plt.text(df['salary_lower'].median() + 1000, plt.ylim()[1] * 0.9, f'Median: ${df["salary_lower"].median()}', color='red')
plt.title('Distribution of Salary Lower Bound')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(x='salary_upper', data=df, kde=True)
plt.axvline(df['salary_upper'].median(), color='red', linestyle='--', label='Median')
plt.text(df['salary_upper'].median() + 1000, plt.ylim()[1] * 0.9, f'Median: ${df["salary_upper"].median()}', color='red')
plt.title('Distribution of Salary Upper Bound')
plt.show()
```

```python
df['text_corpus'] = df['title'].fillna(' ') + ' ' + df['company/description'].fillna('') + ' ' + df['bulletPoints/0'].fillna('') + ' ' + df['bulletPoints/1'].fillna('') + ' ' + df['bulletPoints/2'].fillna('')
df['text_corpus'] = df['text_corpus'].str.lower()
df['text_corpus']
```

```python
from wordcloud import WordCloud

text = ' '.join(df['text_corpus'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Company and Job Descriptions')
plt.show()
```
